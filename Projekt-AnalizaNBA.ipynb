```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("database_24_25.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df_clean = df.dropna(subset=["Player", "PTS", "AST", "TRB"]).copy()
```

```python
df_clean["all_around"] = df_clean["PTS"] + df_clean["AST"] + df_clean["TRB"]
```

```python
player_all = (
    df_clean
    .groupby("Player")["all_around"]
    .mean()
    .sort_values(ascending=False)
)

player_all.head(10)
```

```python
player_all.mean()
```

```python
player_all.min(), player_all.max()
```

```python
player_all.shape[0]
```

```python
top10 = player_all.head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top10.plot(kind="bar")
plt.title("Top 10 najbardziej wszechstronnych zawodnikow NBA (2024/25)")
plt.xlabel("Zawodnik")
plt.ylabel("PTS + AST + TRB (srednia)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

Analiza sprawdza wszechstronność graczy jako zsumowana liczbe punktów, asyst oraz zbiórek. Pozwala ona wyodrebnic 10 zawodników, którzy maja największy wpływ na gre zespołu w kilku aspektach jednoczesnie. Średnia zsumowana wartosc dla calej ligi NBA wynosi okolo 14.23, natomiast najnizsza wartosc w czolowej dziesiatce wynosi 40.85. Dzieki tej analizie można zauważyc bardzo wyraźna różnice pomiędzy przeciętnymi zawodnikami a najlepszymi graczami ligi.
